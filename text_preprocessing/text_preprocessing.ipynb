### html  tag removable

```python
import re
##regular expression to remove html tags
```

```python
def remove_html_tag(text):
    text = re.sub(r"<(script|style)\b[^>]*>.*?</\1>", "", text, flags=re.IGNORECASE | re.DOTALL)
    text = re.sub(r"<[^>]+>", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
remove_html_tag(Text)
```

```python
data["review"]=data["review"].apply(remove_html_tag)
```

```python
data["review"]
```

### url removable technique


```python

def remove_url(text):
    if pd.isna(text):
        return text
    return re.sub(r"https?://\S+|www\.\S+", "", text)

sample_text = "Visit https://example.com and www.testsite.org for more info"
remove_url(sample_text)

```

```python
# apply to the review column
data["review"] = data["review"].apply(remove_url)
data["review"].head()
```

### we gona handdle the punctuation 

```python
import string
```

```python
string.punctuation
```

```python
exclude = string.punctuation
```

```python
for char in exclude:
    print(char)
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, "")
    return text
```

```python
text = "Hello, World!@#$% This is a test."
```

```python
remove_punctuation(text)
```

```python
data["review"] = data["review"].apply(remove_punctuation)
data["review"].head()
```

### genz term ko  handel kerna

```python
text1 ="FYI this is a sample text with numbers 12345 and special characters ."
text2 = "OYO is a popular hotel chain."
text3 = "do this work ASAP"
```

```python
chat_words ={
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "OYO": "On Your Own",
    "B4": "Before",
    "BTW": "By The Way"
}
```

```python
def chat_conversion(text):
    new_text = []
    for word in text.split():
        if word.upper() in chat_words:
            new_text.append(chat_words[word.upper()])
        else:
            new_text.append(word)
    return " ".join(new_text)
```

```python
print(chat_conversion(text1))
```

```python
data["review"] = data["review"].apply(chat_conversion)
data["review"].head()
```
